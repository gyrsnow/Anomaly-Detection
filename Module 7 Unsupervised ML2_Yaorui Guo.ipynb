# About dataset:
This dataset will show you how price for the same diagnosis and the same treatment and in the same city can vary differently across different providers. Variation of hospital charges in the various hospitals in the US for the top 100 diagnoses.

This dataset is owned by the U.S. government. It is available for free on data.gov, and the dataset is updated regularly here.

### Attributes Description
- DRG Definition: The code and description identifying the MS-DRG. MS-DRGs are a classification system that groups similar clinical conditions (diagnoses) and the procedures furnished by the hospital during the stay.
- Provider Id: The CMS Certification Number (CCN) assigned to the Medicare certified hospital facility.
- Provider Name: The name of the provider.
- Provider Street Address: The provider’s street address.
- Provider City: The city where the provider is located.
- Provider State: The state where the provider is located.
- Provider Zip Code: The provider’s zip code.
- Provider HRR: The Hospital Referral Region (HRR) where the provider is located.
- Total Discharges: The number of discharges billed by the provider for inpatient hospital services.
- Average Covered Charges: The provider's average charge for services covered by Medicare for all discharges in the MS-DRG. These will vary from hospital to hospital because of differences in hospital charge structures.
- Average Total Payments: The average total payments to all providers for the MS-DRG including the MSDRG amount, teaching, disproportionate share, capital, and outlier payments for all cases. Also included in average total payments are co-payment and deductible amounts that the patient is responsible for and any additional payments by third parties for coordination of benefits.
- Average Medicare Payments: The average amount that Medicare pays to the provider for Medicare's share of the MS-DRG. Medicare payments DO NOT include beneficiary co-payments and deductible amounts nor any additional payments from third parties for coordination of benefits.

## Section 5.1 Autoencoder Algorithms

## What is Autoencoder
Autoencoder is a common unsupervised learning dimension reduction algorithm based on neural networks, which is used to convert high-dimensional data into low-dimensional data. Autoencoder consists of two main parts: encoder and decoder. encoder is used to compress the original high-dimensional input data, and decoder is used to reconstruct the original high-dimensional input data.

If the new sample is a normal sample, the difference between Output and Input calculated by the model is small, but if it is an abnormal sample, the difference between the model output and the sample will be large, as a sign to determine whether it is an abnormal sample.

## Section 5.2 iForest Algorithms

## What is iForest
iForest is somewhat similar to a random forest, which is composed of multiple itrees and uses random sampling of a portion of data to construct each tree. After the tree is constructed, the prediction process is to search the test records from the root node of the iTree to determine the predicted point to fall on that leaf. 

The assumption of detecting anomalies is that anomalies are generally rare, so they will be quickly assigned to the initial leaf nodes. The anomalies are generally labeled as the leaf node to the root node with a short path in iForest.

## Business Insight
- Both Autoencoder and iForest are very efficient models in detecting anomalies. The proportion of  medical abuse and waste is detected between 0.5% to 1.0%.

- New 2 features average Medicare payments by HRR & DRG are great in detecting anomalies. All 10 features have higher outlier scores in cluster 1 than cluster 2.

- The feature metric is that the median shows a higher outlier score relative to the mean. 

- When I was building Autoencoder, I found that the epochs were within about 20 or even lower and the value loss started to stop changing. Therefore, in order to improve the efficiency of the model, the parameter epochs can be adjusted in the future to reduce the unnecessary model training time.

- When training a tree, an isolated forest randomly draws a certain number of samples from the sample. When there are too many samples (hyper-parameter: max_samples), their distribution in the high-dimensional space will be denser and the tree will be more complex, which is not conducive to outliers being distinguished.

- The question of how to evaluate the performance of the model remains. Whether the outliers go up or down, we have no way of knowing whether the model is performing better or worse. But the idea is to focus specifically on these outliers, where medical abuse and waste can occur.


## Limitation
- Hyper-parameter testing is a very important part of building neural network models. I simply tuned the number of hidden layers and neurons, but the results did not change significantly. If possible, we need to improve the model by tuning the parameters in the future.

- Since the models are constructed relatively few, only the averaging and MOM methods can be used in aggregating the models. If more models can be constructed, the AOM and MOA methods can be tried and there are more chances to find better models.
